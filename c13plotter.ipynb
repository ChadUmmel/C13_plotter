```python
import matplotlib.pyplot as plt
```

```python
mC = 13.00335483521
ma = 4.00260325413
md = 2.01410177811
```

```python
fHaris = open("harissopulos.dat","r")
fDrotleff = open("drotleff.dat","r")
fBrune = open("brune.dat","r")
EHaris = []
csHaris = []
errHaris = []
EDrotleff = []
csDrotleff = []
errDrotleff = []
EBrune = []
csBrune = []
errBrune = []
csBruneadj = []
errBruneadj = []
```

```python
header = fHaris.readline()
header = fHaris.readline()
for line in fHaris:
    entry = line.split()
    EHaris.append(float(entry[0])*1000/4) #Converting to keV/u
    csHaris.append(float(entry[1]))
    if entry[0] == 1.005:
        errHaris.append(float(entry[1])*0.016)
    else:
        errHaris.append(float(entry[1])*0.04) #Error reported as 4% for all values but the measurement at 1.005 MeV
```

```python
header = fDrotleff.readline()
header = fDrotleff.readline()
for line in fDrotleff:
    entry = line.split()
    EDrotleff.append(float(entry[0])*1000/4*(mC+ma)/mC) #Converting to keV/u in lab frame
    csDrotleff.append(float(entry[1])*1000) #Converting to mb
    errDrotleff.append(float(entry[2])*1000)
```

```python
header = fBrune.readline()
header = fBrune.readline()
for line in fBrune:
    entry = line.split()
    EBrune.append(float(entry[0])*1000/2) #Converting to keV/u
    csBrune.append(float(entry[1]))
    csBruneadj.append(float(entry[1])*1e-7)
    errBrune.append(float(entry[2]))
    errBruneadj.append(float(entry[2])*1e-7)
```

```python
plt.errorbar(EDrotleff,csDrotleff,errDrotleff,fmt='.',label = r"$\sigma(\alpha,n)$, Drotleff, $\it{et. al.}$ (1993)")
plt.errorbar(EHaris,csHaris,errHaris,fmt = '.',label = r"$\sigma(\alpha,n)$, Harissopulos, $\it{et. al.}$ (2005)")
plt.errorbar(EBrune,csBrune,errBrune,fmt = '.',label=r"$\sigma(d,n)$, Brune & Kavanagh (1992)")
plt.xlim(0,500)
plt.yscale("log")
plt.xlabel("Laboratory Energy (keV/A)")
plt.ylabel("Cross Section (mb)")
plt.legend();
```

```python
plt.errorbar(EDrotleff,csDrotleff,errDrotleff,fmt='.',label = r"$\sigma(\alpha,n)$, Drotleff, $\it{et. al.}$ (1993)")
plt.errorbar(EHaris,csHaris,errHaris,fmt = '.',label = r"$\sigma(\alpha,n)$, Harissopulos, $\it{et. al.}$ (2005)")
plt.errorbar(EBrune,csBruneadj,errBruneadj,fmt = '.',label=r"$\sigma(d,n) \times 10^{-7}$, Brune & Kavanagh (1992)")
plt.xlim(0,500)
plt.yscale("log")
plt.xlabel("Laboratory Energy (keV/A)")
plt.ylabel("Cross Section (mb)")
plt.legend();
```
